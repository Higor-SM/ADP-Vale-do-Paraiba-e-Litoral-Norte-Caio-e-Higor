```python
import pandas as pd
```

```python
file20 = r'C:\Users\55129\Downloads\Candidatos\2020.csv'
file16 = r'C:\Users\55129\Downloads\Candidatos\2016.csv'

p20 = pd.read_csv(file20, delimiter=';', encoding='ISO-8859-1')
p16 = pd.read_csv(file16, delimiter=';', encoding='ISO-8859-1')

cdd = 'DS_GENERO == "FEMININO" & NM_UE == "CAÇAPAVA" | NM_UE == "JAMBEIRO" | NM_UE == "IGARATÁ" | NM_UE == "JACAREÍ" | NM_UE == "SÃO JOSÉ DOS CAMPOS" | NM_UE == "PARAIBUNA" | NM_UE == "MONTEIRO LOBATO" | NM_UE == "SANTA BRANCA" | NM_UE == "CARAGUATATUBA" | NM_UE == "ILHABELA" | NM_UE == "SÃO SEBASTIÃO" | NM_UE == "UBATUBA"'
```

```python
# Numero de mulheres que se candidataram em 2020 em cada cidade:
p20.query(cdd).groupby('NM_UE').size().sort_values().plot(kind='barh')
```

```python
# Numero de mulheres que se candidataram em 2016 em cada cidade:
p16.query(cdd).groupby('NM_UE').size().sort_values().plot(kind='barh')
```
